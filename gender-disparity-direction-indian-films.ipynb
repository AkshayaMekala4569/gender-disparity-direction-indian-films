## Visualizations


### Pie Chart of Male vs Female Directors

```python
import matplotlib.pyplot as plt

# Count male and female directors
gender_counts = df['Gender'].value_counts()

# Custom colors (feel free to swap!)
colors = ['#2c3e50', '#e74c3c']  # Blue for Male, Pink for Female

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',
        startangle=90, colors=colors, wedgeprops={'edgecolor': 'white'})

plt.title('Gender Distribution of Indian Film Directors')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count directors per region/Industry
industry_counts = df['Industry'].value_counts().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=industry_counts.values, y=industry_counts.index, palette='viridis')

# Add data labels
for i, value in enumerate(industry_counts.values):
    ax.text(value - 1, i, str(value), va='center', ha='right', color='white', fontweight='bold')

plt.title('Distribution of Directors by Industry')
plt.xlabel('Number of Directors')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

### Male vs Female Directors across Industries

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count of male and female directors per region
gender_region_counts = df.groupby(['Industry', 'Gender']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.barplot(data=gender_region_counts, x='Industry', y='Count', hue='Gender', palette='Dark2')

plt.title('Gender Representation by Industry in Indian Cinema')
plt.xlabel('Industry')
plt.ylabel('Number of Directors')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Boxplot of Years Active by Gender

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate IQR for female directors
female_directors = df[df['Gender'] == 'Female']
Q1 = female_directors['Years_Active'].quantile(0.25)
Q3 = female_directors['Years_Active'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
female_outliers = female_directors[female_directors['Years_Active'] > upper_bound]

# Get the x-position of 'Female' in the boxplot
gender_order = df['Gender'].unique().tolist()
female_x = gender_order.index('Female')  # This gives the correct x-position

colors=["magenta","blue"]
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Years_Active', hue='Gender', palette=colors)

# Annotate each female outlier just above its dot
for _, row in female_outliers.iterrows():
    plt.text(x=female_x, y=row['Years_Active'] + 1, s=row['Director_Name'],
             color='black', ha='center', va='bottom', fontsize=9)

plt.title('Career Longevity by Gender')
plt.xlabel('Gender')
plt.ylabel('Years Active')
plt.tight_layout()
plt.show()
```

### Line Chart of Debuts by Gender and Decade

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a 'Decade' column
df['Decade'] = (df['Debute_Year'] // 10) * 10

# Group by Decade and Gender
debut_trends = df.groupby(['Decade', 'Gender']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=debut_trends, x='Decade', y='Count', hue='Gender', marker='o', palette=['#2c3e50', '#e74c3c'], linewidth=3 )

plt.title('Debut Year Trends by Gender')
plt.xlabel('Decade')
plt.ylabel('Number of Directors Debuted')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Spotlight on High-Impact Female Directors

```python
# Filter female directors
female_df = df[df['Gender'] == 'Female']

# Sort by impact: female-centric films and career span
top_female_directors = female_df.sort_values(
    by=['Female_Centric_Films', 'Years_Active'],
    ascending=False
).head(10)

# Display key columns
print(top_female_directors[['Director_Name', 'Debute_Year', 'Years_Active', 'Female_Centric_Films']])
```

```python
top_female = df[df['Gender'] == 'Female'].sort_values(
    by='Female_Centric_Films', ascending=False).head(10)

top_female_melted = top_female.melt(id_vars='Director_Name',
                                    value_vars=['Female_Centric_Films', 'Years_Active'],
                                    var_name='Metric', value_name='Value')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_female_melted, x='Director_Name', y='Value', hue='Metric',
            palette=['#ff69b4', '#9370DB'])

plt.title('Female Directors: Career Span vs Female-Centric Output')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Success Rate By Gender

```python
gender_summary = df.groupby("Gender")[["No_of_Hits", "No_of_Moderates", "No_of_Flops"]].sum()
print(gender_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Step 2: Create heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(gender_summary, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5, linecolor='gray')

# Step 3: Final touches
plt.title("Gender-wise Performance Heatmap")
plt.ylabel("Gender")
plt.xlabel("Performance Category")
plt.tight_layout()
plt.show()
```

```python
Success_Rate_by_Gender=df.groupby("Gender")["Success_Rate(%)"].mean()*100
Success_Rate_by_Gender=Success_Rate_by_Gender.round(2)
print(Success_Rate_by_Gender)
```

```python
import matplotlib.pyplot as plt



# Step 2: Plot doughnut chart
colors = ["#F67280" , "#355C7D"]
plt.figure(figsize=(6, 6))
plt.pie(Success_Rate_by_Gender, labels=Success_Rate_by_Gender.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'width': 0.7})
plt.title('Success Rate by Gender')
plt.tight_layout()
plt.show()
```

### Award Analysis

```python
Awards_by_Gender= df.groupby("Gender")["No_of_Awards"].sum()
print(Awards_by_Gender)
```

```python
import matplotlib.pyplot as plt

# Step 1: Prepare the data
Awards_by_Gender = df.groupby("Gender")["No_of_Awards"].sum()

# Step 2: Plot the lollipop chart
plt.figure(figsize=(6, 3))

# Draw stems
plt.hlines(y=Awards_by_Gender.index, xmin=0, xmax=Awards_by_Gender.values,
           color='#355C7D', linewidth=4)

# Draw markers
plt.plot(Awards_by_Gender.values, Awards_by_Gender.index, 'o',
         color='#F67280', markersize=10)

# Step 3: Add data labels beside the lollipops
for i, value in enumerate(Awards_by_Gender.values):
    plt.text(value + 15, Awards_by_Gender.index[i], str(value),  # ‚Üê shifted right by +1
             va='center', ha='left', fontsize=11, color='black')

# Step 4: Final touches
plt.title('Total Awards by Gender')
plt.xlabel('Number of Awards')
plt.xlim(0, max(Awards_by_Gender.values) * 1.3)
plt.tight_layout()
plt.show()
```

### Number of OTT Projects

```python
gender_counts = df.groupby('Gender')['No_of_OTT_Projects'].sum().reset_index()
print(gender_counts)
```

```python
import matplotlib.pyplot as plt

# Sample data
gender_counts = df.groupby('Gender')['No_of_OTT_Projects'].sum().reset_index()

# Custom colors
colors = {'Male': '#1f77b4', 'Female': '#ff7f0e'}  # You can tweak these!

plt.figure(figsize=(8, 5))
bars = plt.bar(gender_counts['Gender'], gender_counts['No_of_OTT_Projects'],
               color=[colors[gender] for gender in gender_counts['Gender']])

# Add data labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{int(height)}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('Number of OTT Projects by Gender')
plt.xlabel('Gender')
plt.ylabel('No. of OTT Projects')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
